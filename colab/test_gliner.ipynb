```python
# Import
from gliner import GLiNER

# Load a multilingual GLiNER model
# ("gliner-multi" works across many languages, including Romanian)
model = GLiNER.from_pretrained("urchade/gliner_multi_pii-v1")
```

```python
label_map = {
  "NUME_PRENUME": "person's full name",
  "CNP": "13-digit personal numeric code",
  "DATA_NASTERII": "date of birth",
  "SEX": "gender or sex",
  "NATIONALITATE": "nationality",
  "LIMBA_VORBITA": "spoken language",

  "ADRESA": "home address",
  "ADRESA_LUCRU": "workplace address",
  "TELEFON_MOBIL": "mobile phone number",
  "TELEFON_FIX": "landline phone number",
  "EMAIL": "email address",
  "COD_POSTAL": "postal code",

  "ORAS_NASTERE": "city of birth",
  "TARA_NASTERE": "country of birth",

  "PROFESIE": "profession or job title",
  "ACTIVITATE": "field of activity or domain",
  "ANGAJATOR": "employer name",
  "VENIT": "income or salary",

  "STARE_CIVILA": "marital status",
  "EDUCATIE": "education level",

  "IBAN": "IBAN bank account number",
  "CONT_BANCAR": "bank account number",
  "CARD_NUMBER": "credit or debit card number",

  "PASAPORT": "passport number",
  "BULETIN": "identity card number",
  "NUMAR_LICENTA": "license or driver's number",

  "ASIGURARE_MEDICALA": "medical insurance number",
  "GRUPA_SANGE": "blood type",
  "ALERGII": "allergies",
  "CONDITII_MEDICALE": "medical conditions",

  "IP_ADDRESS": "IP address",
  "USERNAME": "username",
  "DEVICE_ID": "device identifier",
  "BIOMETRIC": "biometric data",

  "NUMAR_CONTRACT": "contract number",
  "NUMAR_PLACA": "vehicle license plate",
  "CONT_DIGITAL": "digital account",
  "WALLET_CRYPTO": "cryptocurrency wallet",
  "NUMAR_CONT_ALT": "alternative account number",

  "SEGMENT": "client segment",
  "EXPUS_POLITIC": "politically exposed person status",
  "STATUT_FATCA": "FATCA compliance status"
}

```

```python
# Romanian text (your example)
text = """
Numele meu este Andrei Popescu și locuiesc pe Strada Libertății 45, București.
Adresa mea de email este andrei.popescu@example.ro și numărul de telefon este +40 722 123 456.
M-am născut pe 14.03.1990, iar CNP-ul meu este 1960314330018.
Cardul meu bancar are numărul 4532 1234 5678 9012.
"""

labels = list(label_map.values())

# Run prediction
results = model.predict_entities(text, labels, threshold=0.5)

# Print results
for r in results:
    print(r)
```
